```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

data = [(3, 526), (11, 2320), (32, 516), (42, 641), (64, 474)]
plot_df = spark.createDataFrame(data, ["x", "y"])
```

```python
coordinates = plot_df.collect()
x_vals = [row['x'] for row in coordinates]
y_vals = [row['y'] for row in coordinates]
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_vals, y_vals, color='blue', marker='o')
plt.xlabel("Size in KB")
plt.ylabel("Runtime")
plt.title("Size of csv's and Runtimes")
plt.grid(True)
plt.show()
```
